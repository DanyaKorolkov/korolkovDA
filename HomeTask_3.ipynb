```python
#№1
import numpy as np
import matplotlib.pyplot as plt


url = 'http://www.scipy-lectures.org/_downloads/populations.txt'
data = np.loadtxt(url)

years = data[:,0]
hare = data[:,1]
lynx = data[:,2]
carrots = data[:,3]

plt.plot(years, hare, label = 'hare', color = 'mediumpurple')
plt.plot(years, lynx, label = 'lynx', color = 'sandybrown')
plt.xlabel('Years', fontsize=10)
plt.ylabel('Populations', fontsize=10)
plt.legend()
plt.show()
```

```python
#№2
hare_p = hare/(hare + lynx) * 100
lynx_p= lynx/(hare + lynx) * 100

width = 0.8
plt.bar(years, label = 'hare', color = 'mediumpurple', width=width, height = hare_p)
plt.bar(years, label = 'hare', color = 'sandybrown', width=width, height = lynx_p, bottom = hare_p)

for x, y in zip(years, hare_p):
    plt.text(x, y -10, '%d' %y,  ha='center', va = 'center')
for x, y in zip(years, lynx_p):
    plt.text(x, -y+105, '%d' %y,  ha='center', va = 'center')
```

```python
#№3
plt.figure(figsize=(18,8))

plt.subplot(1, 3, 1)
plt.title("Hare\n Mean = %d\nStd =%d" %( hare.mean(),np.std(hare) ) )
plt.xlabel("Years")
plt.bar(years, hare, color = "mediumpurple")
plt.bar( years.mean(), hare.mean() )
plt.errorbar(years, hare, np.std(hare), marker='^', color = 'pink', fmt='.k', alpha = 0.5)


plt.subplot(1, 3, 2)
plt.title("Lynx\n Mean = %d\nStd =%d" %( lynx.mean(),np.std(lynx) ) )
plt.xlabel("Years")
plt.bar(years, lynx, color = "sandybrown")
plt.bar(years.mean(), lynx.mean())
plt.errorbar(years, lynx, np.std(lynx), marker='^', color = 'pink', fmt='.k', alpha = 0.5)

plt.subplot(1, 3, 3)
plt.title("Carrots\n Mean = %d\nStd =%d" %( carrots.mean(),np.std(carrots) ))
plt.xlabel("Years")
plt.bar(years, carrots, color = "lemonchiffon")
plt.bar(years.mean(), carrots.mean())
plt.errorbar(years, carrots, np.std(carrots), marker='^', color = 'pink', fmt='.k', alpha = 0.5)

plt.show()
```

```python
#№4

plt.plot(years, hare, label = 'hare', color = 'mediumpurple')
plt.plot(years, lynx, label = 'lynx', color = 'sandybrown')
plt.plot(years, carrots, label = 'carrots', color = 'khaki' )

plt.scatter(years[np.argmax(hare)], hare.max(), color = 'r')
plt.scatter(years[np.argmax(lynx)], lynx.max(), color = 'r')
plt.scatter(years[np.argmax(carrots)], carrots.max(), color = 'r')

plt.text(years[np.argmax(hare)] + 0.5, hare.max(), "Max Hare")
plt.text(years[np.argmax(lynx)] + 0.5, lynx.max(), "Max Lynx")
plt.text(years[np.argmax(carrots)] + 0.5, carrots.max(),"Max Carrots")

plt.xlabel('Years', fontsize=10)
plt.ylabel('Populations', fontsize=10)
plt.legend()
plt.show()
```

```python
#№5

plt.title("MaxPopulation Proportion")

leader = np.full((len(years)), "H")
count_h=len(years)
count_l = 0

for i in range(len(years)):
    if lynx[i]>hare[i]:
        leader[i] = "L"
        count_h -= 1
        count_l +=1
        
plt.pie((count_h,count_l), explode= (0, 0.1), labels = ("Hares", "Lynx"), shadow = True, colors = ("mediumpurple", "sandybrown") )
plt.show()

print(leader)
```

```python
#№6

plt.figure(figsize=(18,8))

plt.subplot(1, 3, 1)
plt.title("Hare")
plt.xlabel("Years")
colors_h = ["indigo" if i > 50000 else "mediumpurple" for i in hare]
plt.bar(years, hare, color = colors_h, alpha = 0.6)
plt.plot([years[0], years[len(years) -1]], [50000, 50000], 'k--')


plt.subplot(1, 3, 2)
plt.title("Lynx")
plt.xlabel("Years")
colors_l = ["saddlebrown" if i > 50000 else "sandybrown" for i in lynx]
plt.bar(years, lynx, color = colors_l, alpha = 0.6)
plt.plot([years[0], years[len(years) -1]], [50000, 50000], 'k--')


plt.subplot(1, 3, 3)
plt.title("Carrots")
plt.xlabel("Years")
colors_c = ["khaki" if i > 50000 else "lemonchiffon" for i in carrots]
plt.bar(years, carrots, color = colors_c)
plt.plot([years[0], years[len(years) -1]], [50000, 50000], 'k--')

plt.show()
```

```python
#№7

plt.figure(figsize=(18,8))

plt.subplot(1, 3, 1)
plt.title("Hare")
plt.xlabel("Years")
colors_h = ["indigo" if i in np.partition(hare, 2)[:2]  else "mediumpurple" for i in hare]
plt.bar(years, hare, color = colors_h, alpha = 0.6)



plt.subplot(1, 3, 2)
plt.title("Lynx")
plt.xlabel("Years")
colors_l = ["saddlebrown" if i in np.partition(lynx, 2)[:2] else "sandybrown" for i in lynx]
plt.bar(years, lynx, color = colors_l, alpha = 0.6)



plt.subplot(1, 3, 3)
plt.title("Carrots")
plt.xlabel("Years")
colors_c = ["khaki" if i in np.partition(carrots, 2)[:2] else "lemonchiffon" for i in carrots]
plt.bar(years, carrots, color = colors_c)

plt.show()
```

```python
#№8

plt.figure(figsize=(18,8))

plt.xlabel("Years")
plt.plot(years, np.insert(np.diff(hare), 0, 0), color = "mediumpurple" , alpha = 0.6, label = "Hare chanching")

plt.title("Lynx")
plt.xlabel("Years")
plt.bar(years, lynx, color = "sandybrown", alpha = 0.6, label = "Lynx")

plt.legend()
plt.show()

print("Correlation Coef = \n", np.corrcoef(np.insert(np.diff(hare), 0, 0), lynx))
```

```python
#№9

plt.figure(figsize=(30,8))

width =0.4

plt.subplot(1, 2, 1)
plt.title("Hare vs Carrots")
plt.xlabel("Years")
plt.bar(years, hare, color = "mediumpurple", alpha = 0.6, width = width, label = "Hare")
plt.bar(years + width, carrots, color = "lemonchiffon", width = width, label = "Carrots")
plt.legend()

plt.subplot(1, 2, 2)
plt.title("Hare vs Lynx")
plt.xlabel("Years")
plt.bar(years, hare, color = "mediumpurple", alpha = 0.6, width = width, label = "Hare")
plt.bar(years + width, lynx, color = "sandybrown", alpha = 0.6, width = width, label = "Lynx")
plt.legend()

plt.show()
```
