```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as et
```

```python
tree = et.parse('/content/padgett.xml')
root = tree.getroot()
```

```python
my_data = []
```

```python
for x in root.findall('MetaNetwork'):
  for child in x.find('networks')[0]:
    my_data.append(child.attrib)
```

```python
con = np.zeros(256).reshape(16,16)
for i in range(16):
  con[i] = data.loc[data.source == names[i], 'value']
```

```python
graph = pd.DataFrame(data = con, columns = names, index = names)
graph
```

```python
def page_rank (graph, n_steps, names):

  for i in range(n_steps):
    if (i == 0): 
      iterations = pd.DataFrame( data = [1/len(names)] * len(names), index = names )
    else:
      iterations[i] = 0
      for name in names:
        into = graph.loc[graph[name] != 0, name].index 
        temp = 0
        for j in into:

          temp += iterations[i-1][j] / ( graph.iloc[list(names).index(j)].sum() )

        iterations.iloc[list(names).index(name), i] = temp
        iterations[i][name] = temp

  s = iterations[n_steps-1].sort_values(ascending = False)

  top = []
  for name in names:

    top.append( list(s).index(iterations[n_steps - 1][name]) + 1)

  iterations['top'] = top

  return(iterations)

t = page_rank(graph, 5, names)
```
